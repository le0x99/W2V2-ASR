```python
from prettytable import PrettyTable

def count_parameters(model):
    table = PrettyTable(["Modules", "Parameters"])
    total_params = 0
    for name, parameter in model.named_parameters():
        if not parameter.requires_grad: continue
        param = parameter.numel()
        table.add_row([name, param])
        total_params+=param
    print(table)
    print(f"Total Trainable Params: {total_params}")
    return total_params
```

## LayerNorm
Applies Layer Normalization over a mini-batch of inputs.

$$
        y = \frac{x - \mathrm{E}[x]}{ \sqrt{\mathrm{Var}[x] + \epsilon}} * \gamma + \beta
$$

    The mean and standard-deviation are calculated separately over the last
    certain number dimensions which have to be of the shape specified by
    :attr:`normalized_shape`.
    :math:`\gamma` and :math:`\beta` are learnable affine transform parameters of
    :attr:`normalized_shape` if :attr:`elementwise_affine` is ``True``.
    The standard-deviation is calculated via the biased estimator, equivalent to
    `torch.var(input, unbiased=False)`.


- Input: :math:`(N, *)`
- Output: :math:`(N, *)` (same shape as input)

```python
input.shape
```

```python
# normalize over last 3 dims (default)
m = nn.LayerNorm((5,10,10))
# normalize over last 2 dims (default)
m = nn.LayerNorm((10,10))
# normalize over last dim(default)
m = nn.LayerNorm((10))
```

```python
output = m(input)
```

### Simple Linear Regression

```python
np.random.randn(10,1,2).shape
```

```python
#synthetic data generation
noise_level=0.25
true_a = 1
true_b = 2
x = np.random.rand(100, 1)
y = true_a + true_b * x + np.random.randn(100,1)*noise_level

# Shuffles the indices
idx = np.arange(100)
np.random.shuffle(idx)

# Uses first 80 random indices for train
train_idx = idx[:80]
# Uses the remaining indices for validation
val_idx = idx[80:]

# Generates train and validation sets
x_train, y_train = x[train_idx], y[train_idx]
x_val, y_val = x[val_idx], y[val_idx]

# show
pd.DataFrame([x.flatten(),y.flatten()]).T.plot(x=0,y=1,kind="scatter")
```

```python
# specifiy location of tensor (GPU/CPU)
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Our data was in Numpy arrays, but we need to transform them into PyTorch's Tensors
# and then we send them to the chosen device
x_train_tensor = torch.from_numpy(x_train).float().to(device)
y_train_tensor = torch.from_numpy(y_train).float().to(device)
```

#### A single iteration of GD using gd.

Note:
- batch GD : use large batch with all (N) instances for gradient update
- SGD : use single instances for gradient update
- mini batch : use something inbetween

here we use batch gd.

```python
#random params

# Specify tensors for the parameters, they require the gradient for updates.
a = torch.randn(1, requires_grad=True, dtype=torch.float, device=device)
b = torch.randn(1, requires_grad=True, dtype=torch.float, device=device)

lr = 0.02

# yhat with random params.
yhat = a + b * x_train_tensor
error = y_train_tensor - yhat
loss = (error ** 2).mean()
# No more manual computation of gradients! 
# a_grad = -2 * error.mean()
# b_grad = -2 * (x_tensor * error).mean()
# We just tell PyTorch to work its way BACKWARDS from the specified loss!
loss.backward()
# Let's check the computed gradients...
print(a.grad)
print(b.grad)
# UPDATING the parameters
with torch.no_grad():
    a -= lr * a.grad
    b -= lr * b.grad

# PyTorch is "clingy" to its computed gradients, we need to tell it to let it go...
a.grad.zero_()
b.grad.zero_()
```

```python
# full looop
lr = 0.05
n_epochs = 1_000
# Specify tensors for the parameters, they require the gradient for updates.
a = torch.randn(1, requires_grad=True, dtype=torch.float, device=device)
b = torch.randn(1, requires_grad=True, dtype=torch.float, device=device)

for epoch in range(n_epochs):
    yhat = a + b * x_train_tensor
    error = y_train_tensor - yhat
    loss = (error ** 2).mean()
    loss.backward()
    if epoch%100==0:
        print(loss)
    
    # UPDATING the parameters? Not so fast...
    # We need to use NO_GRAD to keep the update out of the gradient computation
    # Why is that? It boils down to the DYNAMIC GRAPH that PyTorch uses...
    with torch.no_grad():
        a -= lr * a.grad
        b -= lr * b.grad
    
    a.grad.zero_()
    b.grad.zero_()
```

### Linear projection

$y = Ax + b$ (set of lin. eqs.)

Mapping $x \in \mathcal{R}^n \to y \in \mathcal{R}^m$

```python
# x, shape = (4, 1)
x=torch.rand(4,)
x
```

```python
# A, shape = (8, 4)
lin_proj = nn.Linear(in_features=4, out_features=8, bias=True)
A = lin_proj.weight
A
```

```python
b = lin_proj.bias
b
```

```python
# y = Ax + b
y = torch.matmul(A, x) + b
y.shape
```

```python
lin_proj(x) == y
```
